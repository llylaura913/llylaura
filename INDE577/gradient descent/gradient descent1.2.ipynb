```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
 
def lossFunction(x):
    return (x-2.5)**2-1
 
# -1 to 6, there are 140 points 
plot_x = np.linspace(-1,6,141)
# plot_y is lose function 
plot_y = lossFunction(plot_x)
 
plt.plot(plot_x,plot_y)
plt.show()
```

```python
def dLF(theta):
    return derivative(lossFunction, theta, dx=1e-6)
```

```python
theta = 0.0
eta = 0.1
epsilon = 1e-6
while True:
    # what is our gradient degree
    gradient = dLF(theta)
    last_theta = theta
    # moving
    theta = theta - eta * gradient
    # if theta is min
    # as gradient decrease, theta decreases
    # check if the difference meet the requirement 
    if(abs(lossFunction(theta) - lossFunction(last_theta)) < epsilon):
        break
print(theta)
print(lossFunction(theta))
```

```python
def gradient_descent(initial_theta, eta, epsilon=1e-6):
    theta = initial_theta
    theta_history.append(theta)
    while True:
        # what is our gradient degree 
        gradient = dLF(theta)
        last_theta = theta
        # moving
        theta = theta - eta * gradient
        theta_history.append(theta)
        # if theta is min
        # as gradient decrease, theta decreases
        # check if the difference meet the requirement 
        if(abs(lossFunction(theta) - lossFunction(last_theta)) < epsilon):
            break
 
def plot_theta_history():
    plt.plot(plot_x,plot_y)
    plt.plot(np.array(theta_history), lossFunction(np.array(theta_history)), color='red', marker='o')
    plt.show()
```

I am going to prove that learning rate cannot be too small or big. 

By changing learning rate, we see how theta will change:

```python
#learning rate = 0.1
eta=0.1
theta_history = []
gradient_descent(0., eta)
plot_theta_history()
print("Gradient descent search times：",len(theta_history))
```
