# K Means Clustering

    K-means clustering is one of the simplest and popular unsupervised machine learning algorithms.


Clustering: the goal is to group similar instances together into clusters. This is a great tool for data analysis, customer segmentation, recommender systems, search engines, image segmentation, semi-supervised learning, dimensionality reduction, and more.

A cluster refers to a collection of data points aggregated together because of certain similarities.

Youâ€™ll define a target number k, which refers to the number of centroids you need in the dataset. A centroid is the imaginary or real location representing the center of the cluster.

```python
import matplotlib.pyplot as plt  
import numpy as np  
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv ('FEV.csv')
x1 = np.c_[df["Age"]]
x2 = np.c_[df["Hgt"]]
y = np.c_[df["FEV"]]
X = np.column_stack((x1,x2))   
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) 
```

Original:

```python
plt.scatter(x1, x2, c = "red", marker='o', label='FEV')  
plt.xlabel('age')  
plt.ylabel('Hgt')  
plt.legend(loc=2)  
plt.show()
```

Clustering:

```python
estimator = KMeans(n_clusters=3) 
estimator.fit(X_train)     
label_pred = estimator.labels_   
print(estimator.cluster_centers_)
```

Graphing clustering (training): 

```python
x1 = X_train[label_pred == 1]
x2 = X_train[label_pred == 2]
plt.scatter(x1[:, 0], x1[:, 1], c = "green", marker='*', label='label1')  
plt.scatter(x2[:, 0], x2[:, 1], c = "blue", marker='+', label='label2')   
plt.legend(loc=2)  
plt.show()
#testing set
print(estimator.predict(X_test)) 
```

Graphing clustering (testing): 

```python
predict_1=X_test[estimator.predict(X_test) == 1]
predict_2=X_test[estimator.predict(X_test) == 2]
plt.scatter(predict_1[:, 0], predict_1[:, 1], c = "skyblue", marker='*', label='predict1')  
plt.scatter(predict_2[:, 0], predict_2[:, 1], c = "greenyellow", marker='+', label='predict2')  
plt.legend(loc=2)  
plt.show()
```
