```python
import sklearn.linear_model

X = np.c_[df["Age"]]
y = np.c_[df["FEV"]]
# Select a linear model
model = sklearn.linear_model.LinearRegression() # Train the model
model.fit(X, y)
# prediction
X_new = [[30]] 
print(model.predict(X_new)) #if X_new  is 30, the value we will get is 7.09287738
```

```python
from sklearn.model_selection import train_test_split

model = sklearn.linear_model.LinearRegression() # Train the model
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model.fit(x_train, y_train)
print('parameter', model.coef_)
print('intercept', model.intercept_)

y_hat = model.predict(x_test)
print('actual', y_hat[:3])
print('predict', y_test[:3])
```

When we have training and testing datasets, well apply a machine-learning linear regression algorithm to predict the petal length on the basis of its petal width: 

    The data we use is usually split into training data and test data. The training set contains a known output and the model learns on this data in order to be generalized to other data later on. We have the test dataset (or subset) in order to test our models prediction on this subset.

1. using a couple of libraries: sklearn.model_selection to split the training and testing dataset, and matplotlib to draw the regression line.
2. Since weve split our data into x and y, now we can pass them into the train_test_split() function as a parameter along with test_size, and this function will return us four variables. Theyre x_train, y_train, x_test, and y_test.
3. if we pass test_size=0.2, then it means we want to keep 20% of the data of the entire dataset for the testing purposes.

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 15

plt.figure(figsize=(10,6))
plt.scatter(x_train, y_train, c='orange', label='trainning')
plt.scatter(x_test, y_test, c='g', marker='D', label='testing')
plt.plot(X, model.predict(X), 'r-')
plt.legend()
plt.xlabel('Age')
plt.ylabel('FEV')
```

Important Output

MSE(mean squared error)



RMSE(Root Mean Squared Error)



MAE (Mean Absolute Error)



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print('MSE',mean_squared_error(y_test,y_hat))
print('RMSE',np.sqrt(mean_squared_error(y_test,y_hat)))
print('MAE',mean_absolute_error(y_test,y_hat))
print('training: R^2',r2_score(y_train,model.predict(x_train)))
print('testing: R^2',r2_score(y_test,y_hat))

print('training: R^2',model.score(x_train,y_train))
print('testing: R^2',model.score(x_test,y_test))
```

```python
from sklearn.linear_model import LinearRegression

#re-orginizing the data
x1 = np.c_[df["Age"]]
x2 = np.c_[df["Hgt"]]
x3 = np.c_[df["Sex"]]
x4 = np.c_[df["Smoke"]]
y = np.c_[df["FEV"]]
X = np.column_stack((x1,x2,x3,x4))#stack x1, x2, x3, x4
modelRegM = LinearRegression() 
modelRegM.fit(X, y)
yFit = modelRegM.predict(X) 
```

Output: 

```python
print("\nModel2: Y = b0 + b1*x1 + b2*x2 + b3*x3 + b4*x4")
print('intercept: w0={}'.format(modelRegM.intercept_))  
print('parameter: w1={}'.format(modelRegM.coef_))  

print('R2{:.4f}'.format(modelRegM.score(X, y))) 
print('MSE{:.4f}'.format(mean_squared_error(y, yFit)))  
print('MAE{:.4f}'.format(mean_absolute_error(y, yFit)))  
```

```python
#visualizing
fig, ax = plt.subplots(figsize=(8, 6))  
ax.plot(range(len(y)), y, 'b-.', label='Sample')  
ax.plot(range(len(y)), yFit, 'r-', label='Fitting')  
ax.legend(loc='best') 
plt.title('Regression analysis with FEV by SKlearn')
plt.xlabel('factors')
plt.ylabel('FEV')
plt.show()
```
